```python
# construct 2 sin waves one with 100hz and one with 200hz and concatenate them
t = np.linspace(0, 1, 4000)
sin1 = np.sin(2 * np.pi * 200 * t)
sin2 = np.sin(2 * np.pi * 400 * t)
# sin3 = np.sin(2 * np.pi * 400 * t)*0 +1

sin = np.concatenate((sin1,))
```

```python
a= np.fft.rfft(sin)
# a = np.abs(a)
b = np.fft.irfft(a)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = go.Figure()
freqs = np.fft.rfftfreq(len(sin), 1/4000)

fig.add_trace(go.Scatter(x=freqs, y=np.abs(a)))
fig.add_trace(go.Scatter(x=freqs, y=np.angle(a)))

fig.show()
```
